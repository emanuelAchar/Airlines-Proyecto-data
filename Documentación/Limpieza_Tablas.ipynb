# CONEXION Y CARGA DE TABLAS

```python
# importo librerias

import numpy as np
import pandas as pd
import os
```

```python
# establezco la conexion entre el notebook y google drive ('monta' google drive en el entorno de colab)

from google.colab import drive

drive.mount ('/content/drive')
```

```python
# Defino la lista de archivos que quiero leer
file_paths = [
    '/content/drive/My Drive/AIRLINES_PARA_PYTHON/aircrafts_data.csv',
    '/content/drive/My Drive/AIRLINES_PARA_PYTHON/airports_data.csv',
    '/content/drive/My Drive/AIRLINES_PARA_PYTHON/boarding_passes.csv',
    '/content/drive/My Drive/AIRLINES_PARA_PYTHON/bookings.csv',
    '/content/drive/My Drive/AIRLINES_PARA_PYTHON/flights.csv',
    '/content/drive/My Drive/AIRLINES_PARA_PYTHON/seats.csv',
    '/content/drive/My Drive/AIRLINES_PARA_PYTHON/ticket_flights.csv',
    '/content/drive/My Drive/AIRLINES_PARA_PYTHON/tickets.csv']
```

```python
# Creo un diccionario para almacenar los DataFrames con el nombre de cada archivo
dataframes = {}

# Leer cada archivo y almacenar el DataFrame en el diccionario
for file_path in file_paths:
    # Extraer el nombre del archivo sin la extensión y sin el path completo
    file_name = os.path.splitext(os.path.basename(file_path))[0]

    # Leer el archivo CSV y asignarlo al diccionario con el nombre del archivo
    dataframes[file_name] = pd.read_csv(file_path)

# cada archivo CSV se almacena en el diccionario 'dataframes' con su nombre de archivo como clave

for name, df in dataframes.items():
    print(f"DataFrame para {name}:")
    print(df.head(), '\n')
```

# AIRCRAFT_DATA

```python
import json

# Separar la cadena json de la columna model
df_aircrafts = dataframes['aircrafts_data']

# Ver las primeras filas para revisar la estructura
print(df_aircrafts.head())

# Función para extraer el valor de 'en' de la columna model
def extract_model(model_str):
    # Convertir el string en formato JSON a un diccionario
    model_dict = json.loads(model_str.replace("'", '"'))  # Reemplazar comillas simples por dobles para que sea válido JSON
    return model_dict.get('en')  # Extraer el valor de la clave 'en'

# Aplicar la función para limpiar la columna 'model'
df_aircrafts['model'] = df_aircrafts['model'].apply(extract_model)

# Verificar el resultado
print(df_aircrafts.head())

# Guardar el DataFrame limpio en la carpeta de Google Drive
output_file_path = '/content/drive/My Drive/AIRLINES_PARA_PYTHON/aircrafts_data_ordenado1.csv'
df_aircrafts.to_csv(output_file_path, index=False)

print(f"Archivo limpio guardado en: {output_file_path}")
```

```python
# Recargar el archivo limpio en el diccionario con el nombre 'aircrafts_data'
dataframes['aircrafts_data'] = pd.read_csv(output_file_path)

# Verificar que el archivo limpio esté ahora en el diccionario
print("Archivo limpio recargado en el diccionario:")
print(dataframes['aircrafts_data'].head())

# Ver los primeros 5 registros de todos los DataFrames en el diccionario
for name, df in dataframes.items():
    print(f"DataFrame para {name}:")
    print(df.head(), '\n')
```

```python
df_airports = dataframes['airports_data']

# Verificar si las columnas 'airport_name' y 'city' ya han sido transformadas
if isinstance(df_airports['airport_name'].iloc[0], str) and not df_airports['airport_name'].str.startswith("{").any():
    print("Las columnas 'airport_name' y 'city' ya han sido transformadas.")
else:
    # Función para extraer el valor de 'en' de una cadena JSON
    def extract_value(json_str):
        # Convertir el string en formato JSON a un diccionario
        json_dict = json.loads(json_str.replace("'", '"'))  # Reemplazar comillas simples por dobles para que sea válido JSON
        return json_dict.get('en')  # Extraer el valor de la clave 'en'

    # Aplicar la función para limpiar las columnas 'airport name' y 'city'
    df_airports['airport_name'] = df_airports['airport_name'].apply(extract_value)
    df_airports['city'] = df_airports['city'].apply(extract_value)

    # Actualizar el DataFrame limpio en el diccionario
    dataframes['airports_data'] = df_airports

# Verificar el resultado
print(df_airports.head())
```

# FLIGHTS

```python
df_flights = dataframes['flights']
df_flights
```

```python
df_flights.info()
```

```python
# verifico valores nulos
df_flights.isnull().sum()

# no muestra valores nulos
```

```python
# verifico valores faltantes
df_flights.isna().sum()

# no muestra valores nulos ni faltantes, pese a que con .head se ven valores /N
```

```python
# contar los valores /N

# Lista de columnas que quiero analizar
columns_to_check = ['scheduled_departure', 'actual_departure', 'scheduled_arrival', 'actual_arrival']

# Iterar sobre cada columna y contar los valores '\N'
for column in columns_to_check:
    num_N = (df_flights [column] == '\\N').sum()
    print(f"Cantidad de \\N en {column}: {num_N}")
```

```python
# Primero, convierto la columna a string, luego reemplaza '/N' por NaT
df_flights ['actual_departure'] = df_flights ['actual_departure'].astype(str).replace('/N', pd.NaT)
df_flights ['actual_arrival'] = df_flights ['actual_arrival'].astype(str).replace('/N', pd.NaT)

# Después de hacer el reemplazo, convierto a datetime
df_flights['actual_departure'] = pd.to_datetime(df_flights['actual_departure'], format='%Y-%m-%d %H:%M:%S%z', errors='coerce')
df_flights['actual_arrival'] = pd.to_datetime(df_flights['actual_arrival'], format='%Y-%m-%d %H:%M:%S%z', errors='coerce')

# Verificar el resultado
print(df_flights[['actual_departure', 'actual_arrival']])
```

```python
# Rellenar NaT con el valor anterior
df_flights ['actual_departure'] = df_flights ['actual_departure'].ffill()
df_flights ['actual_arrival'] = df_flights ['actual_arrival'].ffill()

df_flights
```

```python
# Contar NaT en la columna 'actual_departure'
nat_count_actual_departure = df_flights ['actual_departure'].isna().sum()
print(f"Número de NaT en 'actual_departure': {nat_count_actual_departure}")
```

```python
# Contar NaT en la columna 'actual_arrival'
nat_count_actual_arrival = df_flights ['actual_arrival'].isna().sum()
print(f"Número de NaT en 'actual_arrival': {nat_count_actual_arrival}")
```

```python
# Eliminar filas con NaT en las columnas 'actual_departure' y 'actual_arrival'

df_flights_cleaned = df_flights.dropna(subset=['actual_departure', 'actual_arrival'])

# Verificar la cantidad de filas después de la eliminación
print(f"Número de filas después de eliminar NaT: {df_flights_cleaned.shape[0]}")
```

```python
# convierto las otras dos columnas de fecha
df_flights['scheduled_departure'] = pd.to_datetime(df_flights['scheduled_departure'], format='%Y-%m-%d %H:%M:%S%z', errors='coerce')
df_flights['scheduled_arrival'] = pd.to_datetime(df_flights['scheduled_arrival'], format='%Y-%m-%d %H:%M:%S%z', errors='coerce')
```
